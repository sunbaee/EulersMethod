## 1. Vital SIR (Euler):

```python
# Importing modules
import matplotlib.pyplot as plot;
import numpy as np;
```

### Implementing method:

```python
# Model function
def SIRModel(b, y, v, initInfected, initSusceptible, populationTime, dt):
    t = np.arange(0, populationTime, dt);
    s, i = np.zeros(len(t)), np.zeros(len(t));
    
    s[0], i[0] = initSusceptible, initInfected;
    for k in range(len(t) - 1):
        s[k + 1] = s[k] - (b*s[k]*i[k] - (1 - s[k])*v)*dt;
        i[k + 1] = i[k] + (b*s[k]*i[k] - (y - v)*i[k])*dt;

    return t, s, i;
```

```python
# Model parameters
b, y, v = 0.12, 0.1, 0;
I = 1e-5;
S = 1 - I;

T = 1000;
dt = 0.1;
```

#### 1.1. Plotting graphics:

```python
t, s, i = SIRModel(b, y, v, I, S, T, dt);
plot.plot(t, s, label="Succetible");
plot.plot(t, i, label="Infected");

plot.legend();
plot.grid();
```

#### 1.2. Infected population and peak:

```python
print("Total Infected:", np.sum(i));
peakInfected  = np.argmax(i);

plot.plot(t, i, c="orange", label="Infected");
plot.scatter(t[peakInfected], i[peakInfected], c="k");
```

#### 1.3. Infected with Mortality:

```python
# Parameters
T = 6400;
```

```python
_, _, i = SIRModel(b, y, v, I, S, T, dt); # No mortality
_, _, iM = SIRModel(b, y, 0.001, I, S, T, dt); # With mortality

plot.plot(t, i, label="Infected");
plot.plot(t, iM, label="infected with Mortality");

plot.title("SIR");

plot.legend();
plot.grid();
```
