```python
# Importing useful libraries
import matplotlib.pyplot as plot;
import numpy as np;
```

```python
# Parameters
b = 0.01; # Infectious parameter
y = 0.01; # Recoveral parameter

N = 99; # Initial population
I = 1 # Number of infected 

T = 30; # Population time
dt = 0.1; # Time step
```

```python
timeRange = np.arange(0, T, dt);

s, i  = np.zeros(len(timeRange)), np.zeros(len(timeRange));

s[0] = N;
i[0] = I;

for j in range(len(timeRange) - 1):
    s[j + 1] = s[j] - b * s[j] * i[j] * dt;
    i[j + 1] = i[j] + b * s[j] * i[j] * dt - y * i[j] * dt;

plot.plot(timeRange, s, label="Susceptible", c="g");
plot.plot(timeRange, i, label="Infected", c="r");

plot.legend();
```
